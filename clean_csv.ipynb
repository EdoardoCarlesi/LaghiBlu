```python
import re
from dateutil import parser

def extract_dates(text):
    """
    Extract dates from a string in multiple formats and return them as datetime objects.
    Supports partial dates (year-month, month-day, etc.).
    """
    # Regex patterns for different date formats
    patterns = [
        r"\b(?:\d{4}-\d{2}-\d{2})\b",   # YYYY-MM-DD
        r"\b(?:\d{4}-\d{2})\b",         # YYYY-MM
        r"\b(?:[A-Za-z]{3,9}-\d{1,2})\b" # MonthName-DD (e.g., Sept-25)
    ]
    
    all_matches = []
    for pat in patterns:
        matches = re.findall(pat, text)
        all_matches.extend(matches)
    
    parsed_dates = []
    for m in all_matches:
        try:
            # Parse flexibly with dateutil
            dt = parser.parse(m, fuzzy=True, default=None)
            parsed_dates.append(dt)
        except Exception:
            pass
    
    return parsed_dates
```

```python
import re

def extract_month(date_list) -> int | None:
        
    months = {
    "gen": 1, "gennaio": 1,
    "feb": 2, "febbraio": 2,
    "mar": 3, "marzo": 3,
    "apr": 4, "aprile": 4,
    "mag": 5, "maggio": 5,
    "giu": 6, "giugno": 6,
    "lug": 7, "luglio": 7,
    "ago": 8, "agosto": 8,
    "set": 9, "settembre": 9,
    "ott": 10, "ottobre": 10,
    "nov": 11, "novembre": 11,
    "dic": 12, "dicembre": 12
    }

    for i, element in enumerate(date_list):
        if isinstance(element, str) and element.isdigit():
            n_mnt = '0'
        elif element == '16set':
            n_mnt = '09'
            return n_mnt
        else:
            mnt = element.lower()
            n_mnt = months[mnt]
            str_mnt = f"{n_mnt:02}" 
            return str_mnt
    
    return n_mnt

def extract_year(text: str) -> int | None:
    """
    Extract a year (1900-2099) from a string.
    Returns the first matching year as an integer, or None if not found.
    """
    pattern = r"\b(19\d{2}|20\d{2})\b"
    match = re.search(pattern, text)
    if match:
        return int(match.group(1))
    return None


def split_string(text: str) -> list[str]:
    """
    Split a string using multiple delimiters: underscore (_), dot (.), or dash (-)
    Returns a list of string parts.
    """
    # Regex: split on _, ., or -
    parts = re.split(r'[_\.-]', text)
    return [p for p in parts if p]  # Remove empty strings if any

def split_letters_numbers(text: str) -> list[str]:
    """
    Split a string into sequences of letters and numbers.
    E.g., "Novembre2022" -> ["Novembre", "2022"]
    """
    # Match consecutive letters or consecutive digits
    parts = re.findall(r'[A-Za-zÀ-ÿ]+|\d+', text)
    return parts


```

```python
import numpy as np

def remove_if_element(df: pd.DataFrame, index, target_type) -> pd.DataFrame:
    """
    Check each row of df, and if the first element is of target_type,
    remove it, shift the rest of the row left, and fill the last column with NaN.
    """
    def shift_row(row):
        if row.iloc[index] == target_type:
            row = row[1:].tolist() + [np.nan]  # shift left and add NaN at the end
        return pd.Series(row, index=df.columns)

    return df.apply(shift_row, axis=1)
```
